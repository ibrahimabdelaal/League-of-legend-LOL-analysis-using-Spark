```python
def prepare_data():
  region='EUW1'
  leagues = ['challengerleagues', 'grandmasterleagues', 'masterleagues', 'DIAMOND', 'PLATINUM', 'GOLD', 'SILVER', 'BRONZE', 'IRON']
  distribution=[1,1,3,19,100,250,299,240,40]    #distrubution of the num of the players among the leagues, this is retireived from https://www.leagueofgraphs.com/rankings/rank-distribution
  j=0
  with open ('summoners2.txt','w') as summ:
    for league in range(9):
      if league <=2: url_leagues = 'https://'+ region + '.api.riotgames.com/lol/league/v4/'+ leagues[league]+ '/by-queue/RANKED_SOLO_5x5?api_key=' + api_key
      if league> 2 :url_leagues='https://'+ region + '.api.riotgames.com/lol/league/v4/entries/RANKED_SOLO_5x5/'+ leagues[league]+ '/III?page=1&api_key=' + api_key
      response = requests.get(url_leagues)
      if response.status_code==200:
        if league <=2: 
           entries = response.json()['entries'] 
        else:
            entries = response.json()
        #summonerNames = [names['summonerName'] for names in entries[0:distribution[league]]]
        for names in entries[0:distribution[league]]:
          summ.write((names['summonerName']) )
          summ.write('\n')
        j+=1
      else :
          league=league-1
          print(j)
          print("status error ",response.status_code)
          time.sleep(1)
```

```python
def summname_to_puuid():

  with open ('summoners2.txt','r') as summ:
     with open ('summPUUID2.txt','w') as summ_id:
        lines=summ.readlines()
        print(lines)
        
        for line in range(len(lines)):
            id=lines[line].strip()
            url_id='https://EUW1.api.riotgames.com/lol/summoner/v4/summoners/by-name/'+id+'?api_key=' + api_key
            response = requests.get(url_id)
            if response.status_code==200:
              id = response.json()['puuid']
              summ_id.write((id))
              summ_id.write("\n")
              time.sleep(1)
            elif response.status_code==429:
               print(line)
               print("status code",response.status_code)
               line=line-1
               
            else:
              print("status code",response.status_code)

```

```python
def get_matchid_from_puid():
  ##this function fetches the match ids give puuid of each sumoner
     Match=[]
     with open ('summPUUID2.txt','r') as summ_id:
        lines=summ_id.readlines()
        print(lines)
        
        for line in range(len(lines)):
            id=lines[line].strip()
            url_id='https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/'+id+'/ids?start=0&count=30&api_key=' + api_key
            response = requests.get(url_id)
            if response.status_code==200:
              Match.append(response.json())
              # summ_id.write((id))
              # summ_id.write("\n")
              time.sleep(1)
            elif response.status_code==429:
               print(line)
               print("status code",response.status_code)
               line=line-1
               
            else:
              print("status code",response.status_code)

        return Match
```

```python
def get_match(match_ids):
  ##this function fetches the match ids give puuid of each sumoner
  with open ('matches2.txt','w') as ma:
    All_matches=[]
    for line in range(len(match_ids)):
        id=match_ids[line]
        url_id='https://europe.api.riotgames.com/lol/match/v5/matches/'+id+ '?api_key=' + api_key
        response = requests.get(url_id)
        if response.status_code==200:
          All_matches.append(response.json())
          ma.write(json.dumps(response.json()))
          ma.write("\n")
          time.sleep(1)
        elif response.status_code==429:
            print(line)
            print("status code",response.status_code)
            line=line-1
            
        else:
          print("status code",response.status_code)

    with open('All_matchesfinal', 'w') as f:
             json.dump(All_matches, f)
```
